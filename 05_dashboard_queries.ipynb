```python
%sql
-- Errors by service

USE CATALOG real_time_projects;
USE SCHEMA realtime_log_analytics_alerting;

SELECT service, SUM(log_count) AS total_errors
FROM gold_log_metrics
WHERE level = 'ERROR'
GROUP BY service
ORDER BY total_errors DESC;

-- Response time trend
SELECT window.start, service, avg_response_time
FROM gold_log_metrics
ORDER BY window.start;
```
