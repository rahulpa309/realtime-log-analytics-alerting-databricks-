```python
import random
import time
import uuid
from datetime import datetime, timezone

OUTPUT_PATH = "/Volumes/real_time_projects/realtime_log_analytics_alerting/project_volume/data/kafka_logs/"
LOGS_PER_SECOND = 5

services = [
    "payment-service",
    "order-service",
    "inventory-service",
    "auth-service"
]

levels = ["INFO", "WARN", "ERROR"]
hosts = ["server-01", "server-02", "server-03"]

messages = {
    "INFO": "Request processed successfully",
    "WARN": "Response time slightly high",
    "ERROR": "Service failure or timeout"
}

print("ðŸš€ Kafka-style Log Generator started...")
print(f"Writing logs to {OUTPUT_PATH}")

while True:
    log_event = {
        "event_id": str(uuid.uuid4()),
        "timestamp": datetime.now(timezone.utc).isoformat(),
        "service": random.choice(services),
        "level": random.choices(levels, weights=[70, 20, 10])[0],
        "message": messages[random.choice(levels)],
        "host": random.choice(hosts),
        "response_time_ms": random.randint(50, 3000)
    }

    df = spark.createDataFrame([log_event])
    df.write.mode("append").json(OUTPUT_PATH)

    time.sleep(1 / LOGS_PER_SECOND)
```
